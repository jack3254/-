```
def buildTrain(train, y):  #幾筆資料為一組 並且標準化
  X_train, Y_train = [], []
  a = 90


  for i in range(0,len(train)//a):
    for j in range(0,135):
      if train.iloc[i*a:(i+1)*a,j].std() != 0:
        train.iloc[i*a:(i+1)*a,j] = (train.iloc[i*a:(i+1)*a,j]-train.iloc[i*a:(i+1)*a,j].mean())/train.iloc[i*a:(i+1)*a,j].std()  
    X_train.append(np.array( train.iloc[i*a:(i+1)*a]  ))
    Y_train.append(y)

  return np.array(X_train), np.array(Y_train)
```

```
def copy(train,a):
  # a = 45 #將資料切成30個一組
  count=0
  d = 90 #平均放置於90欄位
  data = clear_data()
  idx = 0
  x = 0
  for i in range(1,(len(train)//a)+1):
    if len(train)<a:
      print('error')
      break
    while idx < i*d :
      
      if x >= len(train): break
      while int(idx) !=count:
        data.loc[count,:] = 0
        count+=1
      data.loc[int(idx),:] = train.loc[int(x),:]
      idx = idx + d/a
      x = x + i * a/((len(train)//a)*a) 
      count+=1

    while count < i*d :
      data.loc[count,:] = 0
      count+=1 
    idx = i*a
    count = i*a
    
  # data.to_csv('/content/'+test+str(a)+'.csv')  #存檔
  return data
```

```
def buildTrain_2(train):  #幾筆資料為一組 並且標準化
  X_train, Y_train = [], []
  a = 90  #幾筆資料
  
  for i in range(0,len(train)//a):
    for j in range(0,135):
      if train.iloc[i*a:(i+1)*a,j].std() != 0:
        train.iloc[i*a:(i+1)*a,j] = (train.iloc[i*a:(i+1)*a,j]-train.iloc[i*a:(i+1)*a,j].mean())/train.iloc[i*a:(i+1)*a,j].std()  
    X_train.append(np.array( train.iloc[i*a:(i+1)*a]  ))
    # Y_train.append(y)

  return np.array(X_train) # , np.array(Y_train)
```
