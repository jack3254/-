```python
import cv2
import mediapipe as mp
import numpy as np
import os
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#載入model
model = tf.keras.models.load_model('model_webcam.h5')
```

```python
def clear_data():
    data = pd.DataFrame(columns=(
                        #右手
                        'right_hand_x_0','right_hand_y_0',
                        'right_hand_x_4','right_hand_y_4',
                        'right_hand_x_8','right_hand_y_8',
                        'right_hand_x_12','right_hand_y_12',
                        'right_hand_x_16','right_hand_y_16',
                        'right_hand_x_20','right_hand_y_20',

                        'left_hand_x_0','left_hand_y_0',
                        'left_hand_x_4','left_hand_y_4',
                        'left_hand_x_8','left_hand_y_8',
                        'left_hand_x_12','left_hand_y_12',
                        'left_hand_x_16','left_hand_y_16',
                        'left_hand_x_20','left_hand_y_20',

                        'pose_x_0','pose_y_0',# 鼻子
                        'pose_x_15','pose_y_15', #左手
                        'pose_x_16','pose_y_16',#右手
                        ))
    return data
```

```python
def deal_0(train):
#   train = train.iloc[:,1:]
  print(train.shape)
  up=1
  a = 90
  #0~12
  error_0 = 1
  error_12 = 1
  error_24 = 1
  error_26 = 1
  error_28 = 1
  for i in range(0,len(train)):
    if train.iloc[i,0] == 0 and error_0 == 1:
      if i % a == 0:
        up=1
        while up != 0 :
          if up>25:
            error_0 = 0
            print('error_0 = 0')
            print('i=',i)
            break
          if train.iloc[i+up , 0] == 0:
            up += 1
          else : 
            train.iloc[i+up-1,0:12] = train.iloc[i+up,0:12]
            up -= 1

      elif i % a == a-1:
        up=1
        while up != 0 :
          if up>25:
            error_0 = 0
            print('error_0 = 0')
            print('i=',i)
            break
          if train.iloc[i-up , 0] == 0:
            up += 1
          else : 
            train.iloc[i-up+1,0:12] = train.iloc[i-up,0:12]
            up -= 1
      else:
        up=1
        while 1 :
            
          if up>25:
            error_0 = 0
            print('error_0 = 0')
            print('i=',i)
            break
          if train.iloc[i + up , 0] == 0:
            up += 1
          else:
            for j in range(1, up+1):
              train.iloc[i+j-1,0:12] = train.iloc[i-1,0:12] - ( (train.iloc[i-1,0:12]-train.iloc[i+up,0:12]) / (up+1) ) * j
            break
          if (i%a)+up>a-1:
            for j in range(1, up):
              train.iloc[i+j,0:12] = train.iloc[i-1,0:12]
            break
            
    
    #12~24
    if train.iloc[i,12] == 0 and error_12 == 1:
      if i % a == 0:
        up=1
        while up != 0 :
          if up>25:
            error_12 = 0
            print('error_12 = 0')
            print('i=',i)
            break
          if train.iloc[i+up , 12] == 0:
            up += 1
          else : 
            train.iloc[i+up-1,12:24] = train.iloc[i+up,12:24]
            up -= 1

      elif i % a == a-1:
        up=1
        while up != 0 :
          if up>25:
            error_12 = 0
            print('error_12 = 0')
            print('i=',i)
            break
          if train.iloc[i-up , 12] == 0:
            up += 1
          else : 
            train.iloc[i-up+1,12:24] = train.iloc[i-up,12:24]
            up -= 1
      else:
        
        up=1
        while 1 :
          if up>25:
            error_12 = 0
            print('error_12 = 0')
            print('i=',i)
            break
          if train.iloc[i + up , 12] == 0:
            up += 1
          else:
            for j in range(1, up+1):
              train.iloc[i+j-1,12:24] = train.iloc[i-1,12:24] - ( (train.iloc[i-1,12:24]-train.iloc[i+up,12:24]) / (up+1) ) * j
            break
          if (i%a)+up>a-1:
            for j in range(1, up):
              train.iloc[i+j,12:24] = train.iloc[i-1,12:24]
            break
    #24~26
    if train.iloc[i,24] == 0 and error_24 == 1:
      if i % a == 0:
        up=1
        while up != 0 :
          if up>25:
            error_24 = 0
            print('error_24 = 0')
            print('i=',i)
            break
          if train.iloc[i+up , 24] == 0:
            up += 1
          else : 
            train.iloc[i+up-1,24:26] = train.iloc[i+up,24:26]
            up -= 1

      elif i % a == a-1:
        up=1
        while up != 0 :
          if up>25:
            error_24 = 0
            print('error_24 = 0')
            print('i=',i)
            break
          if train.iloc[i-up , 24] == 0:
            up += 1
          else : 
            train.iloc[i-up+1,24:26] = train.iloc[i-up,24:26]
            up -= 1
      else:
        up=1
        while 1 :
          if up>25:
            error_24 = 0
            print('error_24 = 0')
            print('i=',i)
            break
          if train.iloc[i + up , 24] == 0:
            up += 1
          else:
            for j in range(1, up+1):
              train.iloc[i+j-1,24:26] = train.iloc[i-1,24:26] - ( (train.iloc[i-1,24:26]-train.iloc[i+up,24:26]) / (up+1) ) * j
            break
          if (i%a)+up>a-1:
            for j in range(1, up):
              train.iloc[i+j,24:26] = train.iloc[i-1,24:26]
            break

  if error_0 == 0 :
    train.iloc[:,0:12] == 0
  if error_12 == 0 :
    train.iloc[:,12:24] == 0


  
  train.iloc[:,24] = train.iloc[:,24] - 0.5
  train.iloc[:,25] = train.iloc[:,25] - 0.5
  for i in range (0,12):
    train.iloc[:,i*2] = train.iloc[:,i*2] - train.iloc[:,24]
    train.iloc[:,i*2+1] = train.iloc[:,i*2+1] - train.iloc[:,25]
  
  train.iloc[:,26] = train.iloc[:,24] - train.iloc[:,0]
  train.iloc[:,27] = train.iloc[:,25] - train.iloc[:,1]
  
  train.iloc[:,28] = train.iloc[:,24] - train.iloc[:,12]
  train.iloc[:,29] = train.iloc[:,25] - train.iloc[:,13]

            
  return train
```

```python

def buildTrain(train):  #幾筆資料為一組
  X_train, Y_train = [], []
  a = 90
  for i in range(0,len(train)//a):
    X_train.append(np.array(train.iloc[i*a:(i+1)*a]))

  return np.array(X_train)
```

```python
cap = cv2.VideoCapture(0)
# fourcc = cv2.VideoWriter_fourcc(*'XVID')
# out = cv2.VideoWriter('output.avi',fourcc, 30.0, (1920,1080)) 
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,760)
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic


holistic = mp_holistic.Holistic(
    min_detection_confidence=0.5, min_tracking_confidence=0.5)
if not os.path.exists('webcam'): # Check if there exist a directory(folder) to save the files converted from the video
        os.makedirs('webcam')
        print("webcam")
idx = 0

while cap.isOpened():
    success, image = cap.read()
    count = 0    
    if not success:
        print("Ignoring empty camera frame.")
        continue
    else:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = holistic.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
#         mp_drawing.draw_landmarks(
#             image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
#         mp_drawing.draw_landmarks(
#             image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
#         mp_drawing.draw_landmarks(
#             image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
        cv2.imshow('MediaPipe Holistic', image)
        if results.right_hand_landmarks != None or results.left_hand_landmarks != None :
            cv2.imwrite(f"frame{idx}.jpg" , image)    
            idx+=1
        #如果看到手開始寫入資料
        



    
    if idx>89:
        break
    if cv2.waitKey(5) & 0xFF == 27:
        break
cv2.destroyAllWindows()
holistic.close()
cap.release()




file_os = '/webcam/'
data = clear_data()  #清空data
idx = 0
hand_list = [0,4,8,12,16,20]
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(
    min_detection_confidence=0.5, min_tracking_confidence=0.5)
for i in range(90):
    image = cv2.imread(f"frame{i}.jpg")
    
    results = holistic.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

    if results.right_hand_landmarks != None:#將右手21點的座標位置記錄下來
        for i in hand_list:   
            data.loc[idx,'right_hand_x_'+str(i)] = results.right_hand_landmarks.landmark[i].x
            data.loc[idx,'right_hand_y_'+str(i)] = results.right_hand_landmarks.landmark[i].y
    else:
        for i in hand_list:   
            data.loc[idx,'right_hand_x_'+str(i)] = 0
            data.loc[idx,'right_hand_y_'+str(i)] = 0



    if results.left_hand_landmarks != None:#將左手21點的座標位置記錄下來
        for i in hand_list:   
            data.loc[idx,'left_hand_x_'+str(i)] = results.left_hand_landmarks.landmark[i].x
            data.loc[idx,'left_hand_y_'+str(i)] = results.left_hand_landmarks.landmark[i].y
    else:
        for i in hand_list:   
            data.loc[idx,'left_hand_x_'+str(i)] = 0
            data.loc[idx,'left_hand_y_'+str(i)] = 0



    if results.pose_landmarks.landmark[0].visibility > 0.6: #偵測到的pose點 信心度必須要能>0.6的值 不然都裝作沒看到
        data.loc[idx,'pose_x_0'] = results.pose_landmarks.landmark[0].x
        data.loc[idx,'pose_y_0'] = results.pose_landmarks.landmark[0].y
    else:
        data.loc[idx,'pose_x_0'] = 0
        data.loc[idx,'pose_y_0'] = 0
    if results.pose_landmarks.landmark[15].visibility > 0.6: #偵測到的pose點 信心度必須要能>0.6的值 不然都裝作沒看到
        data.loc[idx,'pose_x_15'] = results.pose_landmarks.landmark[15].x
        data.loc[idx,'pose_y_15'] = results.pose_landmarks.landmark[15].y
    else:
        data.loc[idx,'pose_x_15'] = 0
        data.loc[idx,'pose_y_15'] = 0
    if results.pose_landmarks.landmark[16].visibility > 0.6: #偵測到的pose點 信心度必須要能>0.6的值 不然都裝作沒看到
        data.loc[idx,'pose_x_16'] = results.pose_landmarks.landmark[16].x
        data.loc[idx,'pose_y_16'] = results.pose_landmarks.landmark[16].y
    else:
        data.loc[idx,'pose_x_16'] = 0
        data.loc[idx,'pose_y_16'] = 0
    idx += 1  
    if idx >89:
        break
data.to_csv('webcam.csv')
data = deal_0(data)
data = buildTrain(data)
data = np.asarray(data).astype(np.float32)
a=model.predict(data)
y_predict = np.argmax(a,axis=1)
if y_predict == [2]:
    print('哥哥')
if y_predict == [6]:
    print('弟弟')
# print(y_predict)


# data.to_csv('webcam.csv')
# data = deal_0(data)
# data = buildTrain(data)

# data = np.asarray(data).astype(np.float32)
# a=model.predict(data)
# y_predict = np.argmax(a,axis=1)
# print(y_predict)
 
# cv2.destroyAllWindows()
# holistic.close()
# cap.release()

# data.to_csv('webcam.csv') 
```
